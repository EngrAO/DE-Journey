```python
df = spark.read.format('parquet').load("/mnt/retail/bronze")
```

```python
display(df)
```

```python
from pyspark.sql.functions import col,trim,upper,when

df1 = df.filter(
    (col('transactionID').isNotNull())&(col('transactionDate').isNotNull())&(col('customerID').isNotNull()))
```

```python
display(df1)
```

```python
df2 = df1.withColumn('paymentType',upper(trim(col('paymentType')))).withColumn('deviceUsed',upper(trim(col('deviceUsed')))).withColumn('storeregion',upper(trim(col('storeregion'))))
display(df2)
```

```python
df3 = df2.withColumn("TransactionDate", col("TransactionDate").cast("Timestamp")).withColumn("Quantity",col("Quantity").cast("int")).withColumn("Amount",col("Amount").cast("float")).withColumn("Discount", col("Amount").cast("float"))
```

```python
df4 = df3.filter(
    (col("Quantity") > 0) & 
    (col("Amount") > 0) &
    (col("Discount") <= col("Amount"))
)
```

```python
df5 = df4.dropDuplicates()
```

```python
df5.write.mode("overwrite").format("parquet").save("/mnt/retail/silver")
```

```python
silver_df.createOrReplaceTempView("silver_df")
```

```python
%sql
select * from silver_df
```

```python
%sql
select count(distinct transactionid) AS CountofTransaction, round(sum(amount),2) AS Revenue, date(transactiondate)
 from silver_df
group by transactiondate

```

```python
%sql
select paymenttype, round(sum(amount),2) as Revenue
from silver_df
group by paymenttype
```

```python
%sql
select storelocation, round(sum(amount),2) as Revenue
from silver_df
group by storelocation
```

```python
%sql
select customerloyaltylevel, round(sum(amount),2) as Revenue
from silver_df
group by 1
```

```python
%sql
select productcategory, round(sum(amount),2) as Sales
from silver_df
group by 1
order by 2 desc
    
```
